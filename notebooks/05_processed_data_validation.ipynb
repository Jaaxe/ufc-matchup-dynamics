```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load final modeling data
df = pd.read_csv('../data/processed/ufc_modeling_data_final.csv')

# 1. Check for standard distribution
# Z-scores should have a mean of 0 and a standard deviation of 1
plt.figure(figsize=(10, 5))
sns.kdeplot(df['Sig_Str_PM_Z'], label='Striking Pace (Z)', fill=True)
sns.kdeplot(df['Control_Ratio_Z'], label='Grappling Control (Z)', fill=True)
plt.title("Final Feature Distributions (Should be centered at ~0)")
plt.axvline(0, color='black', linestyle='--')
plt.legend()
plt.show()

# 2. Relationship Integrity Check
# High striking volume should negatively correlate with high control even after normalization
plt.figure(figsize=(8, 6))
sns.regplot(x='Sig_Str_PM_Z', y='Control_Ratio_Z', data=df, 
            scatter_kws={'alpha':0.2}, line_kws={'color':'red'})
plt.title("Integrity Check: Volume vs. Control Relationship")
plt.xlabel("Normalized Strikes Per Minute")
plt.ylabel("Normalized Control Time")
plt.show()
```
