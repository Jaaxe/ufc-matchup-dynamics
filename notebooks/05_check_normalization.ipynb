```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load new dataset
df = pd.read_csv('../data/processed/ufc_modeling_data_final.csv')

# 1. Check the Distributions of the New Z-Scores
# They should all be centered around ~0.0 (The Average) with a range of roughly -3 to +3
plt.figure(figsize=(10, 5))
sns.kdeplot(df['Sig_Str_PM_Z'], label='Striking Volume (Z)', fill=True)
sns.kdeplot(df['Takedown_Att_PM_Z'], label='Takedown Attempts (Z)', fill=True)
plt.title("Did Normalization Work? (Distributions should be centered at 0)")
plt.xlabel("Standard Deviations from Division Average")
plt.legend()
plt.show()

# 2. Heavyweights vs Flyweights
# Previously, these boxplots were at different heights. NOW, they should be perfectly aligned at ~0
main_classes = ['Flyweight', 'Heavyweight']
df_subset = df[df['Weight_Class'].isin(main_classes)]

plt.figure(figsize=(8, 5))
sns.boxplot(x='Weight_Class', y='Sig_Str_PM_Z', data=df_subset, palette='coolwarm')
plt.title("Sanity Check: Are Heavyweights and Flyweights aligned now?")
plt.axhline(0, color='red', linestyle='--')
plt.ylabel("Z-Score (Normalized Activity)")
plt.show()
```
