```python
import pandas as pd
import os

# Define file paths
# ".." means go up one level from 'notebooks' to the main folder
raw_fights_path = "../data/raw/raw_fights_detailed.csv"

# Check if file exists
if os.path.exists(raw_fights_path):
    print("File found. Loading data.")
    
    # Load data
    df = pd.read_csv(raw_fights_path)
    
    # Show the first 3 rows and all columns
    pd.set_option('display.max_columns', None)
    display(df.head(3))
    
    # Print the shape (rows, columns)
    print(f"\nDataset Dimensions: {df.shape[0]} fights, {df.shape[1]} columns")

    style_keywords = ['dist', 'clinch', 'ground', 'ctrl', 'time']
    found_cols = [c for c in df.columns if any(keyword in c.lower() for keyword in style_keywords)]

    print(f"Found {len(found_cols)} style-related columns.")
    print("Examples:", found_cols[:10]) # Show first 10 matches
    
else:
    print("File not found.")
```
